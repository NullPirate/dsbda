```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk import pos_tag
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
#Setup & Downloads
nltk.download('punkt', quiet=True)
nltk.download('stopwords', quiet=True)
nltk.download('punkt_tab', quiet=True)
nltk.download('wordnet', quiet=True)
nltk.download('averaged_perceptron_tagger', quiet=True)
nltk.download('averaged_perceptron_tagger_eng', quiet=True)

text = "Natural Language Processing is amazing and it helps computers understand human language"
stop_words = set(stopwords.words('english'))
```

```python
# Tokenization
tokens = word_tokenize(text)
filtered_tokens = [w for w in tokens if w.lower() not in stop_words]

print("--- [1] RAW TOKENS ---")
print(tokens)

print("\n--- [2] CLEANED TOKENS (No Stopwords) ---")
print(filtered_tokens)
```

```python
# Stemming & Lemmatization
ps = PorterStemmer()
lemmatizer = WordNetLemmatizer()

stemmed = [ps.stem(w) for w in filtered_tokens]
lemmatized = [lemmatizer.lemmatize(w) for w in filtered_tokens]

print("\n--- [3] STEMMING (Root Chopping) ---")
print(stemmed)

print("\n--- [4] LEMMATIZATION (Dictionary Base) ---")
print(lemmatized)
```

```python
# POS Tagging (real one, not fake)
tags = pos_tag(tokens)

print("\n--- [5] PART-OF-SPEECH TAGS ---")
print(tags)
```

```python
# Stopwords Removal
stop_words = set(stopwords.words('english'))
filtered_tokens = [word for word in tokens if word.lower() not in stop_words]

print("\nAfter Stopwords Removal:\n", filtered_tokens)
```

```python
# TF-IDF Representation
# Combine processed words back into a string for the vectorizer
processed_str = " ".join(lemmatized)

vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform([processed_str])

print("\n--- [6] TF-IDF FEATURE NAMES ---")
print(vectorizer.get_feature_names_out())

print("\n--- [7] TF-IDF WEIGHTS ---")
print(tfidf_matrix.toarray())
```
