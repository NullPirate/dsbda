```python
import pandas as pd
import numpy as np
df = pd.read_csv("./student_sample_data.csv")
df.head()
```

```python
min_age = df['Age'].min()
print("Minmum Age:", min_age)
max_age = df['Age'].max()
print("Max Age:", max_age)
```

```python
mean_age = df['Age'].mean()
print("Mean Age:", mean_age)

median_age = df['Age'].median()
print("Median Age:", median_age)
```

```python
sd_age = df['Age'].std()
print("Standard deviation of Age:", sd_age)
```

```python
percentage = df['Age'].quantile([0.25, 0.5, 0.75])
print(percentage)
```

```python
grouped = df.groupby("Ethnicity")["Age"]
summary = grouped.agg(['mean', 'median', 'min', 'max', 'std'])
print("\nSummary Statistics of Age grouped by Ethnicity:\n")
print(summary)
```
